```python
import pymysql
```

```python
def getUrl(year):
    domain = "http://my.knu.ac.kr/stpo/stpo/cour/listLectPln/list.action";
    urls = []
    conn = pymysql.connect(host="localhost",
                       user="root",
                       password="gmlakd123",
                       db="knutime",
                       charset="utf8")

    cursor = conn.cursor(pymysql.cursors.DictCursor)

    sql = "SELECT * FROM dept_url"
    cursor.execute(sql)

    rows = cursor.fetchall()

    for row in rows:
        url = domain
        
        depth2_code = row['depth2_code'];
        depth3_code = row['depth3_code'];
        param1 = row['parameter1'];
        param2 = row['parameter2'];
        param3 = row['parameter3'];
        
        url += '?' + param2 + '=' + depth2_code;
        url += '&' + param1 + '=' + year;
        
        if param3 != '':
            url += '&' + param3 + '=' + depth3_code;
        
        urls.append(url)

    conn.close()
    
    return urls
```
